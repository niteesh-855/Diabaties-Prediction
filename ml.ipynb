```python
y = df['Outcome']
x = df.drop('Outcome', axis = 1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
dr = DecisionTreeClassifier()
```

```python
lr.fit(X_train,Y_train)
```

```python
dr.fit(X_train,Y_train)
```

```python
lr.score(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)
```

```python
dr.score(X_train,Y_train)
```

```python
dr.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score 
cross_val_score(dr,X_test,Y_test,cv=5)
```

```python
from sklearn.model_selection import cross_val_score 
cross_val_score(dr,X_train,Y_train,cv=5)
```

```python
from sklearn.model_selection import cross_val_score 
cross_val_score(lr,X_train,Y_train,cv=5)
```

```python
from sklearn.model_selection import cross_val_score 
cross_val_score(lr,X_test,Y_test,cv=5)
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
param_grid = {
    'C' : [0.01,0.1,1,10,100,1000],           #regularisation parameter
    'penalty':['l2','l1','elastinet'],   #penalty types
    'solver':['liblinear']  # OptimizationÂ algorithm
}
```

```python
#Setup GridSearch with cross-val 
grid_search = GridSearchCV(estimator = lr,param_grid=param_grid,cv=5,scoring='accuracy')

#fit the model with the training data
grid_search.fit(X_train,Y_train)

#best hyperparametry found
best_params = grid_search.best_params_
print("Best Parameters:",best_params)
```
